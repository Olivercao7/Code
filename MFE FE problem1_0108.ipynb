```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from itertools import product
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/Users/mac/Desktop/combined_data.csv", skiprows=None)
df.set_index('Unnamed: 0').rename_axis('Date')
```

```python
# In sample model-selection
true_false = [True, False]
params = [true_false] * 6
choices = list(product(*params))

#Run regression
factors = sm.add_constant(df.iloc[:, -6:])
y = df[['Momentum', 'Mkt-RF']]
best_Ins_models = {}
best_Ins_fitted_models = {}
for column in y:
    best_Ins_aic = float('inf')
    best_Ins_model = None
    best_Ins_fitted_model = None
    for choice in choices:
        sel = [True] + list(choice)
        x = factors.loc[:, sel]
        res = sm.OLS(y[column],x).fit()
        if res.aic < best_Ins_aic:
            best_Ins_aic = res.aic
            best_Ins_model = x.columns
            best_Ins_fitted_model = res
    best_Ins_models[column] = best_Ins_model
    best_Ins_fitted_models[column] = best_Ins_fitted_model
    print(f"For {column}, AIC selects {best_Ins_model}")
```

```python
res = sm.OLS(y['Momentum'], factors[['const', 'INDPRO', 'DEFAULT']]).fit()
print("For Momentum, AIC selects('const', 'INDPRO', 'DEFAULT')")
print(res.summary())
```

```python
res = sm.OLS(y['Mkt-RF'], factors[['const', 'UNRATE', 'INDPRO','INFLATION']]).fit()
print("For Mkt-RF, AIC selects('const', 'UNRATE', 'INDPRO','INFLATION')")
print(res.summary())
```

```python
def oos_rsquared(y, yhat, mu):

    numerator = ((y - yhat) ** 2).sum()
    denominator = ((y - mu) ** 2).sum()

    r2 = 1 - (numerator / denominator)

    return r2
```

```python
#Out of sample regression anlysis
# 数据划分
# Split data into training and testing sets
middle_index = len(df) // 2
train = df.iloc[:middle_index]
test = df.iloc[middle_index:]
df = sm.add_constant(df)

# Assume the last 4 columns are predictors, 'Momentum', 'Mkt-RF' are response variables
x_train = sm.add_constant(train.iloc[:, -6:])
y_train = train[['Momentum', 'Mkt-RF']]
x_test = sm.add_constant(test.iloc[:, -6:])
y_test = test[['Momentum', 'Mkt-RF']]

# Model selection on the training set
true_false = [True, False]
params = [true_false] * 6
choices = list(product(*params))
best_models = {}
best_fitted_models = {}

for column in y_train:
    best_aic = float('inf')
    best_model = None
    best_fitted_model = None
    for choice in choices:
        sel = [True] + list(choice)
        selected_x_train = x_train.iloc[:, sel]
        res = sm.OLS(y_train[column], selected_x_train).fit()
        if res.aic < best_aic:
            best_aic = res.aic
            best_model = selected_x_train.columns
            best_fitted_model = res
    best_models[column] = best_model
    best_fitted_models[column] = best_fitted_model

# Store the output for each model
# Evaluate the models on the test set and compute out-of-sample R^2
output = {}
for column, best_model in best_models.items():
    diff_y_forecast = []
    diff_y_samplemean = []
    for i in range(middle_index,len(df)):
        selected_factors = df.iloc[i][best_model].values
        y_t = df.iloc[i][column]
        x_train = df.iloc[:i][best_model]
        y_train = df.iloc[:i][column]
        oos_model = sm.OLS(y_train,x_train).fit()
        #Make predictions on the test set
        yhat = oos_model.predict(selected_factors)
        #Compute the in-sample mean from the training set
        mu = y_train.mean()
        #Compute the elements for r^2
        diff_y_forecast.append((y_t - yhat)**2)
        diff_y_samplemean.append((y_t - mu)**2)
    oos_r2 = 1 - (sum(diff_y_forecast)/sum(diff_y_samplemean))
    output[column] = {
        "Best AIC Model": list(best_model),
        "Out-of-Sample R^2": oos_r2
    }
    


output
```

```python
def calculate_weights(expected_return, r_bar):
    return 1.5 if expected_return > r_bar else 0.5
```

```python
Insample_sharpe_ratios = {}
for column, best_aic_combination in best_Ins_models.items():
    Insmodel_predictions = best_Ins_fitted_models[column].predict(x_test[list(best_aic_combination)])
    Insmodel_expected_returns[column] = Insmodel_predictions

for column in y_test.columns:
    portfolio_returns = []
    r_bar = df[column].mean()  # 计算历史回报率的均值

    for i in range(middle_index, len(df)):
        expected_return = Insmodel_expected_returns[column][i]  # 预期回报率
        weight = calculate_weights(expected_return, r_bar)
        current_return = df.iloc[i][column]  # 当前时期的回报率
        rf_return = df.iloc[i]['RF']  # 当前时期的风险无关利率

        portfolio_return = weight * current_return + (1 - weight) * rf_return
        portfolio_returns.append(portfolio_return)

    # 计算夏普比率
    average_return = np.mean(portfolio_returns)
    std_dev = np.std([r - rf for r, rf in zip(portfolio_returns, df.iloc[middle_index:]['RF'])])
    sharpe_ratio = (average_return - np.mean(df.iloc[middle_index:]['RF'])) / std_dev

    Insample_sharpe_ratios[column] = sharpe_ratio

print("Ins_sharpe_ratios:", sharpe_ratios)
```

```python
oos_sharpe_ratios = {}
#计算预期回报率
oosmodel_expected_returns = {}
for column, best_model in best_models.items():
    oos_predictions = best_fitted_models[column].predict(x_test[best_model])
    oosmodel_expected_returns[column] = oos_predictions
for column in y_test.columns:
    portfolio_returns = []
    r_bar = df[column].mean()  # 计算历史回报率的均值

    for i in range(middle_index, len(df)):
        oosmodel_expected_return = oosmodel_expected_returns[column][i]  # 预期回报率
        weight = calculate_weights(oosmodel_expected_return, r_bar)
        current_return = df.iloc[i][column]  # 当前时期的回报率
        rf_return = df.iloc[i]['RF']  # 当前时期的风险无关利率

        portfolio_return = weight * current_return + (1 - weight) * rf_return
        portfolio_returns.append(portfolio_return)

    # 计算夏普比率
    average_return = np.mean(portfolio_returns)
    std_dev = np.std([r - rf for r, rf in zip(portfolio_returns, df.iloc[middle_index:]['RF'])])
    sharpe_ratio = (average_return - np.mean(df.iloc[middle_index:]['RF'])) / std_dev

    oos_sharpe_ratios[column] = sharpe_ratio

print("oos_sharpe_ratios:", sharpe_ratios)
```
